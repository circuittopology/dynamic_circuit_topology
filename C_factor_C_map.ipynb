**READ C FACTOR FILES**

*Choose if you want to analyze C or C* factor (C_factor/Cstar_factor)* 

```python
#Read C factor (or C* factor) data and calculate average over the runs
c_factor='Cstar_factor'
path_Cfactor='results/C_factor'
path_figures= 'results/figures/all protein'
runs=['run1', 'run2', 'run3']
savefig= 1
C_factor_data=create_df_cfactor(path_Cfactor, runs, c_factor)
avg=(C_factor_data['run1']+C_factor_data['run2']+C_factor_data['run3'])/len(runs)
```

```python
#Make comprehensive plot of C factor over all three runs
left, width = 0.1, 1.1
bottom, height = 0.1, 0.13
spacing = 0.005
params= [left, width,bottom, height,spacing]

f= plot_rectangle(avg, C_factor_data, params, path_figures, c_factor, savefig)
```
